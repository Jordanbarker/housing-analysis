## The state of the economy

The bond 3-month, 10-year Treasury spread refers to the difference between the interest rates on the 10-year U.S. Treasury note and the 3-month U.S. Treasury bill. Historically, an inverted curve (when the 3-month yield is above the 10-year yield) has been a reliable predictor of recessions. When the 3-month yield exceeds the 10-year yield, the spread is negative, known as a **yield curve inversion**. This often signals that investors expect short-term economic trouble or a recession, as they are seeking the safety of long-term bonds over short-term instruments.
